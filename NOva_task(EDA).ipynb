```python
import pandas as pd
```

```python
df=pd.read_csv("TASK.csv")
```

```python
df.isnull().sum()/len(df)
```

```python
df.drop(columns=["storm_name","time","injuries","location_description","photos_link"],inplace=True)
```

```python
df.isnull().sum()/len(df)
```

```python
df.fillna(method="ffill",inplace=True)
```

```python
df.isnull().sum()/len(df)
```

```python
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')


df['tstamp'] = pd.to_datetime(df['tstamp'], errors='coerce')


categorical_cols = ['country', 'nearest_places', 'hazard_type', 'landslide_type', 'trigger',
                    'source_name', 'source_link', 'location_accuracy', 'landslide_size',
                    'cat_src', 'countryname', 'near', 'adminname1', 'adminname2',
                    'countrycode', 'continentcode', 'key']
df[categorical_cols] = df[categorical_cols].fillna('Unknown')


numerical_cols = ['fatalities', 'distance', 'population', 'version', 'changeset_id', 'latitude', 'longitude']
df[numerical_cols] = df[numerical_cols].fillna(0)


df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.histplot(df['fatalities'], bins=30, kde=True)
plt.title('Distribution of Fatalities')
plt.xlabel('Number of Fatalities')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['population'], bins=30, kde=True)
plt.title('Distribution of Population Affected')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(x='population', y='fatalities', data=df)
plt.title('Fatalities vs Population Affected')
plt.xlabel('Population')
plt.ylabel('Fatalities')
plt.show()


plt.figure(figsize=(12, 8))
sns.boxplot(x='continentcode', y='fatalities', data=df)
plt.title('Fatalities by Continent')
plt.xlabel('Continent')
plt.ylabel('Fatalities')
plt.show()


plt.figure(figsize=(12, 6))
sns.countplot(x='landslide_type', data=df, order=df['landslide_type'].value_counts().index)
plt.title('Count of Landslide Types')
plt.xlabel('Landslide Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(x='longitude', y='latitude', data=df, hue='continentcode', palette='viridis', legend='full')
plt.title('Geographical Distribution of Events')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster

# Initialize the map centered around the average coordinates
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=2)

# Add points to the map
marker_cluster = MarkerCluster().add_to(m)

for idx, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"{row['country']} - {row['hazard_type']}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(marker_cluster)

# Display the map
m.save("map.html")
```

```python
import plotly.express as px

# Create a scatter plot on a map
fig = px.scatter_geo(df,
                     lat='latitude',
                     lon='longitude',
                     hover_name='country',
                     hover_data=['hazard_type', 'fatalities'],
                     color='continentcode',
                     title='Geographical Distribution of Events')

fig.update_geos(projection_type="natural earth")
fig.update_layout(height=600, margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
